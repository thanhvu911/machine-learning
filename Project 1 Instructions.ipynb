## Preprocessing Pipeline
Encapsulate your final preprocessing steps (scaling numeric features, encoding categorical features) into a reusable function or scikit-learn Pipeline. I should be able to create a clean dataset by running a single line of code. This might look like:
- `clean_data = pipeline.fit_transform(raw_data)`
- `clean_data = my_cleaning_fn(raw_data)`

**Your preprocessing steps should include:**
-  Convert categorical variables into dummy variables.
-  Scale numeric features . *Justify your choice if not using Standard Scaler.*
-  Handling missing data and outliers, as decided in the previous section.

**After preprocessing, check your work:**
-  Show the final shape of your clean dataset.
-  Print the first few rows to make sure all transformations worked.
-  Generate summary statistics for all features.

```python
# your code goes here!
def preprocess_data(raw_df):
    df = raw_df.drop(columns=['game_name'])
    
    numeric_features = ['price_usd', 'current_players', 'peak_today', 'user_rating', 'num_reviews']
    categorical_features = ['developer', 'primary_genre','release_year']
    
    num_imputer = SimpleImputer(strategy='mean')
    df[numeric_features] = num_imputer.fit_transform(df[numeric_features])
    
    cat_df = pd.get_dummies(df[categorical_features])
    
    scaler = StandardScaler()
    num_scaled = scaler.fit_transform(df[numeric_features])
    num_df = pd.DataFrame(num_scaled, columns=numeric_features, index=df.index)
    
    clean_df = pd.concat([num_df, cat_df], axis=1)
    
    return clean_df

clean_data = preprocess_data(df)

print("Final shape of clean dataset:", clean_data.shape)
print("First few rows:")
print(clean_data.head())
print("Summary statistics:")
print(clean_data.describe())
```
